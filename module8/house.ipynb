## Linear Regression
A method to help us understand the relationship between two variables:
- The predictor/independent variable (X)
- The response/dependent variable (Y)

THe result of a linear regression is a function that predicts

$
\widehat{Y} = a + bX
$

- a referece to `intercept`
- b referce to `slope`



```python
from sklearn.linear_model import LinearRegression
```

```python
# Create a linear regression object

lm = LinearRegression()
lm
```

## Multiple linear regressions
If you want multiple predictors in the model

$
\widehat{Y} = a+ b_{1}X_{1} + b_{2}X_{2} +  b_{3}X_{3} + .. +  b_{n}X_{n}
$

For this example good predictors of price will be:
- sqft_living
- bedrooms
- bathrooms
- sqft_lot

```python
import pandas as pd
from sklearn.linear_model import LinearRegression
df = pd.read_csv('../data/kc_house_data.csv')
```

```python

Z = df[['sqft_living', 'bedrooms', 'bathrooms', 'sqft_lot']]
Y = df['price']
```

```python
lm = LinearRegression()
model_multiple  = lm.fit(Z,Y)
a = model_multiple.intercept_
bs = list(model_multiple.coef_)

print(f'Intercept: {a}, b1: {bs[0]}, b2:{bs[1]}, b3: {bs[2]}, b4: {bs[3]}')
```

```python
Yhat_multiple = a + bs[0]*df['sqft_living'] + bs[1]*df['bedrooms'] + bs[2]*df['bathrooms'] + bs[3]*df['sqft_lot']
```

You need to do a quantitative measure to determine how accurate your model is. THere are two measurements we do:
- $R^{2}$ : Coefficient of determination( how close to the fitted regression line)
- $MeanSquareError(MSE)$ : Measure the average of the square error
