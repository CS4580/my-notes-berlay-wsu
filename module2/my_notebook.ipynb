One of the most common magic commands is the `%mathplotlib inline` command. This command will enable plots to be displayed inside the notebook.

````python

%matplotlib inline
````

````python
""" Simple demo of scatter plot"""

import numpy as np
import matplotlib.pyplot as plt

n =50
x = np.random.rand(n)
y = np.random.rand(n)

colors = np.random.rand(n)
# calculate the area
area = np.pi*(15*np.random.rand(n))**2 # 0 to 15 point radius
#plot data
plt.scatter(x,y,s=area, c=colors, alpha =0.5)
plt.show()
````

### Python list vs Numpy Arrays
- Elements of different type `vs` same type
- Variable size nesting is possible `vs` same size
