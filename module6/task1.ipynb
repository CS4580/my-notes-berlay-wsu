## Task: Game platform preference
Yo are interested in creating your own video game. Analyze a data set from IGN which ranks the different platforms and game types.

```python
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns
print('Setup Complete')
```

### Load and review analysis

For this analysis we will us the: `data/ing_scores.scv`
- Use the Platform column as index.

```python
data_path = '../data/m6/ign_scores.csv'

# df = pd.read_csv(path2, index_col='Platform')
# df.head()

game_df = pd.read_csv(data_path, index_col='Platform')
game_df.head()
```

### Which platform is the best?
Create a bar chart that shows the average score for `racing` games, for each platform.
if `Wii` platform is not the best, which platform is?

```python
# use bar plot
plt.figure(figsize=(15,6))
plt.title('Platform ratings')

sns.barplot(x=game_df.index, y=game_df['Racing'])
plt.show()
```

```python
# code forr thwe bar plot
platform_high_score = game_df['Racing'].idxmax()
print(f'Platform with the highest score for racing is {platform_high_score}')
```

### Task: All possible combinations
Create a heatmap of the average scores by genre in each cell

```python
# Plot heatmap for all data

plt.figure(figsize=(15,6))

plt.title('Heatmap')
sns.heatmap(data=game_df, annot=True)
plt.xlabel('Genre')
plt.show()
```
