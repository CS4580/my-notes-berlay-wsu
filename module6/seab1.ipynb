```python
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns
import numpy as np
```

```python
data_path = '../data/seabor/museum_visitors.csv'
museum_df = pd.read_csv(data_path, index_col='Date',parse_dates=True)
museum_df.head()
```

- Create a line plot of vistors `Avila Abobe`
- Replace `inf` (missing data) with `nan`, and then fill them in with 0

```python
# replace infinite values with NaN
museum_df = museum_df.replace( [np.inf, -np.inf], np.nan)

# now we can use the fillna() method replace NaN with zeros
museum_df = museum_df.fillna(0)
```

```python
# plot tha data
plt.figure(figsize=(16.6))
sns.lineplot(museum_df['Avila Adobe'])
```

### Bar Plot
Dataset: US Department of Transportation that tracks flight delays


```python
# load file
data_path = '../data/seabor/flight_delays.csv'
flight_df = pd.read_csv(data_path, index_col='Month',parse_dates=True)
flight_df.head()
```

```python
# use bar plot
plt.figure(figsize=(15,6))
plt.title('FLight delay')

sns.barplot(x=flight_df.index, y=flight_df['NK'])
plt.show()
```

### HEATMAPS

```python
# Plot heatmap for all data

plt.figure(figsize=(15,6))

plt.title('Heatmap')
sns.heatmap(data=flight_df, annot=True)
plt.xlabel('Airline')
plt.show()
```
