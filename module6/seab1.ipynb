```python
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns
import numpy as np
```

```python
data_path = '../data/m6/museum_visitors.csv'

museum_df = pd.read_csv(data_path, index_col='Date',parse_dates=True)
museum_df.head()
```

- Create a line plot of vistors `Avila Abobe`
- Replace `inf` (missing data) with `nan`, and then fill them in with 0

```python
# replace infinite values with NaN
museum_df = museum_df.replace( [np.inf, -np.inf], np.nan)

# now we can use the fillna() method replace NaN with zeros
museum_df = museum_df.fillna(0)
```

```python
# plot tha data
plt.figure(figsize=(16.6))
sns.lineplot(museum_df['Avila Adobe'])
```

### Bar Plot
Dataset: US Department of Transportation that tracks flight delays


```python
# load file
data_path = '../data/m6/flight_delays.csv'
flight_df = pd.read_csv(data_path, index_col='Month')
flight_df.head()
```

```python
# use bar plot
plt.figure(figsize=(15,6))
plt.title('FLight delay')

sns.barplot(x=flight_df.index, y=flight_df['NK'])
plt.show()
```

### HEATMAPS

```python
# Plot heatmap for all data

plt.figure(figsize=(15,6))

plt.title('Heatmap')
sns.heatmap(data=flight_df, annot=True)
plt.xlabel('Date')
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns
print('Libraries loaded')
```

```python
data_path = '../data/m6/insurance.csv'

insurance_df = pd.read_csv(data_path)
insurance_df.head()
```

Create a `scatter` plot. In seaborn us `sns.scatterplot` command
- x-axis use `bmi`
- y-axis use `charges`

```python
sns.scatterplot(insurance_df, x='bmi', y='charges')
plt.xlabel='bmi'
plt.ylabel='charges'
plt.show()
```

### Plot suggest we have a `positive correlation` between bmi and changes. 
- People with higher bmi pays more.

To double check the strength of this correlation , add `regression line` that best fir the data

```python
sns.regplot(x=insurance_df['bmi'], y=insurance_df['charges'])
plt.show()
```

### Color-coded scatter plots
They are good to display correlation between more than two variables.

We want to understand how 'smoker` are affects the relationship

```python
# Scatter plot of bmivs charges and soker vs chages
# x= bmi, y=chargers, hue=smokers

sns.scatterplot(insurance_df, x='bmi', y='charges', hue='smoker')
plt.show()
```

```python
# for regression lines

sns.lmplot(insurance_df, x='bmi', y='charges', hue='smoker')
plt.show()
```

Finally, we can adapt the design of the scatter plot to feature a `categorical variables` (like `smoker`) on one of the main axis. Some people call this plot: **Categorical Scatter Plot**

```python
sns.swarmplot(insurance_df, x='smoker',y='charges')
plt.show()
```
