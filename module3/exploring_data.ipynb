# Using iloc (index location) to select data

Format: `df_data.iloc[row_indexer, column_indexer]`

```python
# starts with col 0 until 4th rows for all columns
df_yeast.iloc[0:4]
```

```python
# provide rows and columns
df_yeast.iloc[0:4, 2:5]
```

```python
# Select rows using bolling indexer (valid under condition)
# get all record whos column 9 = 'MIT'

df_yeast[df_yeast[9]=='MIT']
```

```python
# Another

df_yeast[df_yeast[1]>= 0.5]
```

```python
#replace values
# Assign 'Waldo Weber' to col_0, for all rows that match col_9 = 'MIT'

df_yeast.iloc[df_yeast[9] =='MIT', 0] = 'Waldo Weber'
df_yeast
```

```python
# to understand the data set use `describe` method
df_yeast.describe()
```

```python
# get the min(), max(), and std from col_04
print(f'Min: {df_yeast[4].min()}')
print(f'Max: {df_yeast[4].max()}')
print(f'Standard deviation: {df_yeast[4].std()}')
```
