## Maps

- Take one set of values and maps to another set
- Use it when you want to represent the data in a different way.

There are two common `map()` methods: 
- `map()` : to transform on one value/column
- `apply()` : to transform the whole dataframe

```python
reviews_point_mean = reviews_d.points.mean()
```

```python
# now map the values
reviews_d.points.map(lambda p: p-reviews_point_mean)
```

The `map()` should expect single value from a `series` (a point value), and returns a transformed version of a value.
`map()` returns a new series where all values have been transformed

Research: First class citizen and comprehensions.

```python
def remean_points(row):
    row.points = row.points - reviews_point_mean
    return row

# Call it
reviews_d.apply(remean_points, axis='columns') # In python can pass the function as parameter. "First Class Citizen"

# print first  records

reviews_d.head()
```
