## Scatter plot

## Bar plot

Useful when you are showing or comparing segments of information, or categorical data

```python
# bar plot
city = pd.read_csv('../data/cities.csv')
city.head()
```

```python
# filter records from Year=2019

city_19 = city[city['Year']==2019]
city_19.head()
```

```python
# plot a bar chart number of cities

city_19.plot(kind='bar', x ='City', y = 'Population', title="Population of cities in 2019", figsize=(10,6))
plt.show()
```

```python
# you may want to plot data horizontally
# since we plot only one set of values we don't need legend
city_19.plot(kind='barh', x='City', y='Population', title='Population of cities in 2019 horizontally', legend=False, figsize=(10,6))
plt.show()
```

### Scatter plots

When you are interested in the relationship between two variables, scatter plots are the best choice.


```python
# Load data
houses = pd.read_csv('../data/homes.csv')
houses.head()
```

```python
houses.plot(kind='scatter', x ='List', y='Rooms')
plt.show()
```

```python
# Filtering data directly
houses[houses['Age']<50].plot(kind='scatter', x='Age',y='List')
plt.show()
```

```python
# Eliminate outliers, and filter
house_plot_df=houses[houses['Age']<50]
house_plot_df=house_plot_df[house_plot_df['List']<350]

# now plot
house_plot_df.plot(kind='scatter', x='Age', y='List',title='Age less than 50 and Price<350')
plt.show()
```

### Area plot
Useful when want to understand how values develop and aer composed over time

```python
enrol = pd.read_csv('../data/enrollment.csv')
enrol.head()
```

```python
# Create are plot of the enrollments

enrol.plot(kind='area', x='Year')
plt.show()
```

```python
# improve the viz by organizing the data by year
# we want to see the total. and how it changes over time
# sort the data by year, then plot it using pivot method

pivot_df = enrol.pivot(index='Year', columns='Class', values='Enrollment')
pivot_df.head()
```

```python
pivot_df.plot(kind='area')
plt.show()
```

```python
# The data is stacked by default. 
# You can turn off this option
pivot_df.plot(kind='area', stacked=False)
plt.show()
```
